Why we need Pandas?

In a dataset, there are different columns representing features. Pandas manages tabular data more efficiently. 

What can we do with Pandas?

* Import datasets (CSV, Excel, SQL)
* Data cleansing (Missing values, Invalid values)
* Size mutability (Add / Delete rows/columns)
* Reshaping and pivoting of datasets
* Efficient manipulation and extraction
* Statistical analysis

Two primary data structures available in Pandas: 

* Series (1-Dimensional - Represents a single column) - A one-dimensional labeled homogeneous array, size immutable. The columns items (each row) is identified by an index.
* Dataframe (2-Dimensional) - A two-dimensional labeled, size-mutable tabular structure with heterogeneously types columns.

Data structure are collections of data types that provide the best way of organizing item (or values) in terms of memory usage.

loc function - label based indexing

Label Based Indexes - Here, you start and end values -both are included.

```python
import pandas as pd
import numpy as np

s = pd.Series([10, 20, 30, 40, 50])
s.name = "Calories"

index = ["apple", "banana", "grapes", "orange", "strawberry"]

s.index = index

print(s)

print(s.loc['banana'])

print(s['banana':'orange'])

print(s.loc[['grapes', 'apple']])
```

Logical Operators: AND / OR / NOT

```python
import pandas as pd
import numpy as np

# grams of protein

fruit_protein = {
    "Avocado": 2.0,
    "Guava": 2.6,
    "Blackberries": 2.0,
    "Oranges": 0.9,
    "Banana": 1.1,
    "Apples": 0.3,
    "Kiwi": 1.1,
    "Pomegranate": 1.7,
    "Mango": 0.8,
    "Cherries": 1.0
}

s2 = pd.Series(fruit_protein, name="Protein")

print(s2[(s2>0.5) & (s2<=2)]) # AND operation

print(s2[(s2>0.5) | (s2<=2)]) # OR operation

print(s2[~(s2>1)]) # NOT operation
```

Filling Missing Values - fillna()

```python
import pandas as pd
import numpy as np

data3 = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve", "Alice"],
    "Age": [25, 30, 35, np.nan, 29, 25],
    "Department": ["HR", "IT", "Finance", "IT", "HR", "HR"],
    "Salary": [50000, 60000, 70000, 62000, np.nan, 50000],
    "Promoted Salary": [500000, 600000, 700000, 620000, np.nan, 500000]
}

df = pd.DataFrame(data3)

print(df.fillna(0))

print(df)
```

```python
# Filling the missing values in the Age column with the mean value

import pandas as pd
import numpy as np

data3 = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve", "Alice"],
    "Age": [25, 30, 35, np.nan, 29, 25],
    "Department": ["HR", "IT", "Finance", "IT", "HR", "HR"],
    "Salary": [50000, 60000, 70000, 62000, np.nan, 50000],
    "Promoted Salary": [500000, 600000, 700000, 620000, np.nan, 500000]
}

df = pd.DataFrame(data3)

print(df["Age"].fillna(df["Age"].mean())) # Here, "inplace = True" is not used.

print(df)
```

```python
# Filling the missing values in the Salary column with the median value

import pandas as pd
import numpy as np

data3 = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve", "Alice"],
    "Age": [25, 30, 35, np.nan, 29, 25],
    "Department": ["HR", "IT", "Finance", "IT", "HR", "HR"],
    "Salary": [55000, 65000, 75000, 67000, np.nan, 55000],
}

df = pd.DataFrame(data3)

print(df["Salary"].fillna(df["Salary"].median())) # Here, "inplace = True" is not used.

print(df)
```
