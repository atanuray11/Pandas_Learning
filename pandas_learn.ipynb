Why we need Pandas?

In a dataset, there are different columns representing features. Pandas manages tabular data more efficiently. 

What can we do with Pandas?

* Import datasets (CSV, Excel, SQL)
* Data cleansing (Missing values, Invalid values)
* Size mutability (Add / Delete rows/columns)
* Reshaping and pivoting of datasets
* Efficient manipulation and extraction
* Statistical analysis

Two primary data structures available in Pandas: 

* Series (1-Dimensional - Represents a single column) - A one-dimensional labeled homogeneous array, size immutable. The columns items (each row) is identified by an index.
* Dataframe (2-Dimensional) - A two-dimensional labeled, size-mutable tabular structure with heterogeneously types columns.

Data structure are collections of data types that provide the best way of organizing item (or values) in terms of memory usage.

Indexing

```python
import pandas as pd
import numpy as np

s = pd.Series([10, 20, 30, 40, 50])

print(s[0])

print(s[0:2]) # start (included) end (not included) step value (values to jump)
print(s[2:4])
```

iloc - Location based indexing

```python
import pandas as pd
import numpy as np

s = pd.Series([10, 20, 30, 40, 50])

print(s.iloc[3])

print(s.iloc[[1, 3, 4]])
```

```python
import pandas as pd
import numpy as np

s = pd.Series([10, 20, 30, 40, 50])
s.name = "Calories"

index = ["apple", "banana", "grapes", "orange", "strawberry"]

s.index = index

print(s)

print(s['grapes'])

# Here, iloc cannot be used by refereing the names of the indexes. It can still be used by referengin the index number.
print(s.iloc[3])
```

loc function - label based indexing

Label Based Indexes - Here, you start and end values -both are included.

```python
import pandas as pd
import numpy as np

s = pd.Series([10, 20, 30, 40, 50])
s.name = "Calories"

index = ["apple", "banana", "grapes", "orange", "strawberry"]

s.index = index

print(s)

print(s.loc['banana'])

print(s['banana':'orange'])

print(s.loc[['grapes', 'apple']])
```

Creating a Series using Dictionary

```python
import pandas as pd
import numpy as np

# grams of protein

fruit_protein = {
    "Avocado": 2.0,
    "Guava": 2.6,
    "Blackberries": 2.0,
    "Oranges": 0.9,
    "Banana": 1.1,
    "Apples": 0.3,
    "Kiwi": 1.1,
    "Pomegranate": 1.7,
    "Mango": 0.8,
    "Cherries": 1.0
}

s2 = pd.Series(fruit_protein, name="Protein")

print(s2)
```

Conditional Selection

```python
import pandas as pd
import numpy as np

print(s2>1)
```

```python
import pandas as pd
import numpy as np

print(s2[s2>1])
```

Logical Operators: AND / OR / NOT

```python
import pandas as pd
import numpy as np

# grams of protein

fruit_protein = {
    "Avocado": 2.0,
    "Guava": 2.6,
    "Blackberries": 2.0,
    "Oranges": 0.9,
    "Banana": 1.1,
    "Apples": 0.3,
    "Kiwi": 1.1,
    "Pomegranate": 1.7,
    "Mango": 0.8,
    "Cherries": 1.0
}

s2 = pd.Series(fruit_protein, name="Protein")

print(s2[(s2>0.5) & (s2<=2)]) # AND operation

print(s2[(s2>0.5) | (s2<=2)]) # OR operation

print(s2[~(s2>1)]) # NOT operation
```

Modifying the Series

```python
import pandas as pd
import numpy as np

# grams of protein

fruit_protein = {
    "Avocado": 2.0,
    "Guava": 2.6,
    "Blackberries": 2.0,
    "Oranges": 0.9,
    "Banana": 1.1,
    "Apples": 0.3,
    "Kiwi": 1.1,
    "Pomegranate": 1.7,
    "Mango": 0.8,
    "Cherries": 1.0
}

s2 = pd.Series(fruit_protein, name="Protein")

s2["Mango"] = 2.8

print(s2)
```
